```python
import streamlit as st
import torch
from diffusers import StableDiffusionPipeline, EulerDiscreteScheduler

# Model Details
model_id = "stabilityai/stable-diffusion-2"

# Initialize the scheduler
scheduler = EulerDiscreteScheduler.from_pretrained(model_id, subfolder="scheduler")

# Initialize the pipeline with the scheduler
pipe = StableDiffusionPipeline.from_pretrained(model_id, scheduler=scheduler, torch_dtype=torch.float16)

# Move the pipeline to the GPU (if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
pipe = pipe.to(device)

def generate_image(prompt):
    image = pipe(prompt, num_inference_steps=50, guidance_scale=7.5).images[0]
    return image

def main():
    st.title("Stable Diffusion 2 Model")
    st.write("Generate an image based on your text prompt.")

    prompt = st.text_area("Enter your prompt", height=200)

    if st.button("Generate"):
        image = generate_image(prompt)
        st.image(image, caption="Generated Image")

if __name__ == "__main__":
    # Run the Streamlit app in a Colab environment
    import subprocess
    subprocess.run(["streamlit", "run", "streamlit_app.py"])
```
